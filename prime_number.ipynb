```
### 素数算出 ###
### 約数の検証をその数の平方根より小さい素数に限定
### 算出された数字は保証しません。小林

import numpy as np

max_num = 1000
pri_num = [2]

def descri(n):
  for _ in pri_num:
    if (_ > int(np.sqrt(n)) + 1):
      break
    if (n % _) == 0:
      return(0)
      break
  return(1)

# 素数を抽出
for i in range(2, max_num):
  if descri(i) == 1:
    pri_num.append(i)

# 10個毎に改行して表示
print("Prime number")
j = 1
for i in pri_num:
  if j % 10 == 0:
    print("{:,}".format(i), ", ")
  else:
    print("{:,}".format(i),", ", end="")
  j += 1
```
